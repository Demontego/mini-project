```python
from bs4 import BeautifulSoup
import re
import pandas as pd
from selenium import webdriver
import html
from time import sleep
from html.parser import HTMLParser
import random
import numpy as np
import requests
import selenium.common.exceptions as selexcept
from multiprocessing.dummy import Pool as ThreadPool
from multiprocessing import Lock, Value
from datetime import datetime
from selenium.webdriver.common.by import By
```

```python
def random_sleep(offset=1.5, length=4):
    sleep(random.random() * length + offset)
games=[]
games.clear()
count=0
with webdriver.Edge() as driver:
    driver.get('https://store.steampowered.com/search/?filter=topsellers')
    random_sleep()
    try:
        while True:
            items = driver.find_element_by_id('search_results').find_elements_by_tag_name('div')[2].find_elements_by_tag_name('a')
            for item in items:
                games.append(item.get_attribute('href'))
            print(count, len(games))
            link = driver.find_element_by_class_name('search_pagination_right').find_element_by_partial_link_text('>')
            print("link found")
            link.click()
            sleep(1)
            count+=1
    except selexcept.NoSuchElementException:
        pass
```

```python
def extract_game_info(game_url):
    game_info={}
    cookies={'Steam_Language': 'russian'}
    game_html = requests.get(game_url, cookies=cookies).text
    soup = BeautifulSoup(game_html, 'html.parser')
    
    
    name=soup.find('div', class_='apphub_AppName')
    if(name):
        game_info['Название']=name.text
    else:
        name=soup.find('h2', class_='pageheader')
        if(name):
            game_info['Название']=name.text
        else:
            return game_info
    
    price=soup.find('div', class_='game_purchase_price')
    if(price):
        try:
            game_info['цена']=float(re.findall('\d+', price.text)[0])
        except:
            game_info['цена']=0
        game_info['конечная_цена']=game_info['цена']
        game_info['скидка']=0
    else:
        bp=soup.find('div', class_='discount_original_price')
        if(bp):
            bp=bp.text.replace(',','.')
            game_info['цена']=float(re.findall('\d+', bp)[0])
    
        discount=soup.find('div', class_='discount_pct')
        if(discount):
            game_info['скидка']=int(re.findall('\d+', discount.text)[0])
        
        ap=soup.find('div', class_='discount_final_price')
        if(ap):
            ap=ap.text.replace(',','.')
            game_info['конечная_цена']=float(re.findall('\d*', ap)[0])
    
    block=soup.find('div', class_='block responsive_apppage_details_left')
    if block:
        mp=block.find('a', class_="name",href="https://store.steampowered.com/search/?category2=1&snr=1_5_9__423")
        if(mp):
            game_info['Мультиплеер']=1.0
        else:
            game_info['Мультиплеер']=0.0
        
        sp=block.find('a', class_="name",href="https://store.steampowered.com/search/?category2=2&snr=1_5_9__423")
        if(sp):
            game_info['Одиночный']=1
        else:
            game_info['Одиночный']=0
        
        coop=block.find('a', class_="name",href="https://store.steampowered.com/search/?category2=9&snr=1_5_9__423")
        if(coop):
            game_info['Кооператив']=1
        else:
            game_info['Кооператив']=0
        
        MMO=block.find('a', class_="name",href="https://store.steampowered.com/search/?category2=20&snr=1_5_9__423")
        if(MMO):
            game_info['ММО']=1
        else:
            game_info['ММО']=0
        
    table=soup.find('table', class_='game_language_options')
    if(table):
        rows = table.find_all('tr')
        data = [list(map(lambda x: x.text.strip(), row.find_all('td',class_="ellipsis"))) for row in rows]
        game_info['Русский']= int(['русский'] in data)
        game_info['Англикий']=int(['английский'] in data)
        game_info['количество_языков']=len(data)-1
        
    positive_recent=soup.find('span',class_="responsive_hidden")
    if(positive_recent):
        game_info['недавние_оценки+']=int(re.sub(r'[,()]', '', positive_recent.text.strip()))
        
    marks=soup.find_all('span', class_='user_reviews_count')
    if(marks):
        game_info['все_оценки']=int(re.sub(r'[,()]', '',marks[0].text))
        game_info['позитивная_оценка']=int(re.sub(r'[,()]', '',marks[1].text))
        game_info['негативная_оценка']=int(re.sub(r'[,()]', '',marks[2].text))

    genre=soup.find('div',class_="details_block")
    if(genre):
        genre=genre.find('a')
        if(genre):
            game_info['жанр']=genre.text
    
    game_data = requests.get(game_url).text
    soup_data = BeautifulSoup(game_data, 'html.parser')
    date=soup_data.find('div', class_='date')
    if(date):
        date=date.text.replace(',','').strip()
        try:
            game_info['дата_релиза']=datetime.strptime(date, '%d %b %Y').date()
        except:
            pass
        
    pub=soup.find('div', class_="user_reviews")
    if(pub):
        pub=pub.find_all('a')
        try:
            game_info['Издатель']=pub[1].text
        except:
            pass
        
        
    systems=soup.find('div', class_='game_area_sys_req_rightCol')
    if systems:
        try:
            items=systems.find_all('li')
            if items:
                for item in items:
                    item=item.text.split(':')
                    if(len(item)<2):
                        continue
                    game_info[item[0].strip()]=item[1].replace('//', ',').strip()
        except:
            pass
    systems=soup.find('div', class_='game_area_sys_req_full')
    if systems:
        items=systems.find_all('li')
        if items:
            for item in items:
                item=item.text.split(':')
                if(len(item)<2):
                    continue
                game_info[item[0].strip()]=item[1].replace('//', ',').strip()

    
    return game_info
```
